
Preprocessing


```python
import numpy as np
def preprocess_input(df):
  df=df.copy()
  df=df.query("Element == 'Temperature change'")
  #group contries and take their mean
  df=df.groupby('Area').mean()

  #use only time series data
  df=df.loc[:,'Y1961':]

  #get global mean of temperature change
  df=pandas.DataFrame(df.mean()).reset_index(drop=False)
  df.columns=['ds','y']

  #fix year column
  df['ds']=df['ds'].apply(lambda x:x[1:]).astype(np.int)
  return df
```

```python

time_series=preprocess_input(data)
```

In- Sample Forecast


```python
time_train=time_series.iloc[:44,:].copy()
time_test=time_series.iloc[44:,:].copy()
```

```python
time_test
```

```python
in_model=Prophet()
in_model.fit(time_train)
```

```python
in_forecast=in_model.predict(time_test).loc[:,['ds','yhat']]
in_forecast['ds']=in_forecast['ds'].apply(lambda x: x.year)
in_forecast
```

```python
in_result_df=time_series.merge(in_forecast, on='ds',how='left')
fig=px.line(
    in_result_df,
    x='ds',
    y=['y','yhat'],
    color_discrete_sequence=['black','red'],
    labels={'ds':"year"},
     title="Average Global Temperature Change over time"
   
)
fig.show()
```

```python
import numpy as np
future_df=pandas.DataFrame(np.arange(2020,2046),columns=['ds'])
future_df
```
